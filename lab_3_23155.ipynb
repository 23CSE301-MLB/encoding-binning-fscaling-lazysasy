```python
df4 = df.copy()

from sklearn.preprocessing import MinMaxScaler
numeric_cols = df4.select_dtypes(include='number').columns

scaler = MinMaxScaler()
scaled_data = scaler.fit_transform(df4[numeric_cols])

scaled_df = pd.DataFrame(scaled_data, columns=numeric_cols)

df4[numeric_cols] = scaled_df

print(df4.head())
```

```python
df5 = df.copy()

from sklearn.preprocessing import StandardScaler
numeric_cols = df5.select_dtypes(include='number').columns

scaler = StandardScaler()
scaled_data = scaler.fit_transform(df5[numeric_cols])

scaled_df = pd.DataFrame(scaled_data, columns=numeric_cols)

df5[numeric_cols] = scaled_df

print(df5.head())
```

```python
df6 = df.copy()

numeric_cols = df6.select_dtypes(include='number').columns
imputer = SimpleImputer(strategy='mean')
df6[numeric_cols] = imputer.fit_transform(df6[numeric_cols])

normalizer = Normalizer(norm='l2')
normalized_data = normalizer.fit_transform(df6[numeric_cols])

normalized_df = pd.DataFrame(normalized_data, columns=numeric_cols)

df6[numeric_cols] = normalized_df
print(df6.head())
```

```python
age_data = df[['Age']].dropna()
age_data['age_bins_equal'] = pd.cut(age_data['Age'], bins=4, labels=['Young', 'Adult', 'Middle-aged', 'Senior'])

print(age_data[['Age', 'age_bins_equal']].head())
```

```python
age_data['age_bins_quantile'] = pd.qcut(age_data['Age'], q=4, labels=['Q1', 'Q2', 'Q3', 'Q4'])

print(age_data[['Age', 'age_bins_quantile']].head())
```

```python
custom_bins = [0, 30, 50, 70, 100]
custom_labels = ['Youth', 'Young Adult', 'Middle Age', 'Senior']

age_data['age_bins_custom'] = pd.cut(age_data['Age'], bins=custom_bins, labels=custom_labels)
print(age_data[['Age', 'age_bins_custom']].head())
```
